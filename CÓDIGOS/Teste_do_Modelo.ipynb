```python
import tensorflow as tf
import numpy as np
import os
from tensorflow.keras.preprocessing import image

PASTA_COM_IMAGENS = './pasta_teste'
CAMINHO_MODELO = 'modelo_tumor_final.keras'

def testar_tudo():
    if not os.path.exists(CAMINHO_MODELO):
        print("ERRO: Modelo n칚o encontrado.")
        return

    print("Carregando IA...")
    model = tf.keras.models.load_model(CAMINHO_MODELO)

    arquivos = os.listdir(PASTA_COM_IMAGENS)

    contagem_tumor = 0
    contagem_normal = 0

    print(f"\n{'ARQUIVO':<30} | {'PREVIS츾O':<15} | {'CONFIAN칂A'}")
    print("-" * 65)

    for arquivo in arquivos:

        if arquivo.startswith('.'):
            continue

        if not arquivo.lower().endswith(('.png', '.jpg', '.jpeg')):
            continue

        caminho_completo = os.path.join(PASTA_COM_IMAGENS, arquivo)

        try:

            img = image.load_img(caminho_completo, target_size=(224, 224))
            img_array = image.img_to_array(img)
            img_array = np.expand_dims(img_array, axis=0)

            score = model.predict(img_array, verbose=0)[0][0]

            if score > 0.5:
                resultado = "游댮 TUMOR"
                confianca = score * 100
                contagem_tumor += 1
            else:
                resultado = "游릭 NORMAL"
                confianca = (1 - score) * 100
                contagem_normal += 1

            print(f"{arquivo:<30} | {resultado:<15} | {confianca:.1f}%")

        except:

            continue

    total_validos = contagem_tumor + contagem_normal

    print("-" * 65)
    print("\n游늵 RESUMO FINAL:")
    print(f"Total Analisado:      {total_validos}")
    print(f"Tumores detectados:   {contagem_tumor}")
    print(f"Normais detectados:   {contagem_normal}")

if __name__ == "__main__":
    testar_tudo()
```
